```python
from flask import Flask,render_template,request
import pickle
import numpy as np

app=Flask(__name__)

# Handle missing model.pkl by using a dummy model
try:
    model=pickle.load(open('model.pkl','rb'))
except FileNotFoundError:
    class DummyModel:
        def predict(self, X):
            # A simple linear model for demonstration if model.pkl is missing
            # Assumes X is a numpy array like [[experience]]
            if X.shape == (1,1):
                experience = X[0,0]
                # Example: salary = 5000 * experience + 30000
                return np.array([[5000 * experience + 30000]])
            else:
                return np.array([[0.0]]) # Default for unexpected input shape

    model = DummyModel()
    print("Warning: 'model.pkl' not found. Using a dummy model for prediction.")

@app.route('/')
def home():
    return render_template('home.html')

@app.route('/predict',methods=['POST'])
def predict():
    exp=float(request.values['experience'])
    exp=np.reshape(exp,(-1,1))
    output=model.predict(exp)
    output=output.item()
    output=round(output,2)
    return render_template('res.html',prediction_text="You are eligible for a salary of Rs.{}".format(output))

if __name__ == '__main__':
    app.run(port=8000)
```
